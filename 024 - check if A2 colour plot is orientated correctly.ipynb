```python
def trace_over_angle(angle_degrees, plot = False,divisions = None):
    '''This function is expected to return the energy gap traced over a path defined by an angle theta'''
    theta = angle_degrees/180. * np.pi
    xv, yv = np.meshgrid(np.linspace(0,1,gaps.shape[0]), np.linspace(0,1,gaps.shape[0]))
    r =(xv**2+yv**2)**0.5
    data = np.transpose(gaps.copy())
    data[r>1]=0
    x = np.linspace(0, 1, gaps.shape[0])*np.cos(theta)*gaps.shape[0]
    y = np.linspace(0, 1, gaps.shape[0])*np.sin(theta)*gaps.shape[0]
    gap_path = scipy.ndimage.map_coordinates(data, np.vstack((x,y)))

    if plot:
        plt.figure()
        plt.plot(np.linspace(0, 1, gaps.shape[0]),gap_path)
        plt.xlabel('1-s')
        plt.ylabel('$\Delta \lambda $')
        plt.xlim((0,1))
        plt.ylim((0,max(gap_path)))
        plt.show()
        return
    else:
        return gap_path
```

```python
# this cell is particular to the data loaded at the top
h_i = DATA['centre_run'].h
J_ij = DATA['centre_run'].J
qubits = DATA['centre_run'].qubits
if ver==1:
    split = [['log','joining'],['centre']]
elif ver==2:
    split = [['log'],['centre','joining']]
N = (qubits+5)/3#Q = 3N-5
M1 = N-3 # there are N-3 joining bits
M2 = M1+1 # there are M1+1 centre bits
Q=M1+M2+N# = 3N-5 # Total qubits

sfB = mylib.linear_schedule
def A_sched(sfl,sfa,split):
    ranges = {'log':range(N),'joining':range(N,N+M1),'centre':range(Q-M2,Q)}
    A_sched =  [0]*Q
    for i in range(2):
        for j,sett in enumerate(split[i]):
            for k in ranges[sett]:
                A_sched[k]=[sfl,sfa][i]
    return A_sched

def test_single_run(theta):
    '''This function just does an annealing run at the defined angle'''
    sfl = lambda points: np.cos(theta)*mylib.linear_schedule(points)
    sfa = lambda points: np.sin(theta)*mylib.linear_schedule(points)
    sched_funcs = [A_sched(sfl,sfa,split),[sfB]]
    a = mylib.Anneal(qubits,[h_i,J_ij],T=3.,points = 1000,sched_funcs = sched_funcs, diff_scheds = True, show_bar = True, light = False) 
    a.run()
    return a

```

```python
def compare_path_and_run(angle_degrees):
        '''This function uses prevous functions so I can make sure that the trace and a new run with the same defining
        angle return the same result'''

        theta = (angle_degrees/180.)*np.pi
        a = test_single_run(theta)
        if CONSIDER_DEGERERATE_SOLUTIONS: 
            ind1,ind2 = 7,8
        else: 
            ind1,ind2 = 0,1
        gap_path = trace_over_angle(angle_degrees, plot = False)
        plt.figure()
        plt.plot(1-a.ss,a.eigenvals[:,ind2]-a.eigenvals[:,ind1],label = 'run')
        plt.plot(a.ss,gap_path,label = 'path')
        plt.ylabel('$\Delta \lambda _{%s,%s}$'%(ind1,ind2))
        plt.xlabel('1-s')
        plt.legend(framealpha = 0, loc = 'best')
        plt.show()
```

```python
# the following function is intended to take an array of data and a series of divisions
# it will find the points where the data crosses any of these divions
def crosser(x,y,divisions):
    x_crosses = []
    for d in divisions:
        x_crosses.append([x[i] for i in np.where(np.diff(np.sign(y-d)))[0]])
    return x_crosses
```
